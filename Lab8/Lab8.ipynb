```python
def swap(mat,ele1, ele2):

  mat1 = []
  for i in mat:
    mat1.append(i)

  temp = mat1[ele1[0]][ele1[1]] 
  mat1[ele1[0]][ele1[1]] = mat1[ele2[0]][ele2[1]]
  mat1[ele2[0]][ele2[1]] = temp

  return mat1
```

```python
def solve1(start):

  parents = {}
  
  states = []
  goalstates = [[[1,2,3], [4,5,6], [7,'a','b']], [[1,2,3], [4,5,6], [7,'b','a']]] 
  frontier = deque()
  explored = set()

  newnode = Node1()
  newnode.state = start
  newnode.todo = 'a'

  frontier.append(newnode)

  # count = 0

  parents[tuple(map(tuple, start))] = None

  while(len(frontier)>0):

    # count+=1
    
    # if(count == 5):
    #   return parents


    currnode = frontier.popleft()
    currt = tuple(map(tuple, currnode.state))
    
    key = currnode.todo

    if currt in explored:
      continue 

    states.append(currnode)

    if currnode.state in  goalstates:
      return currnode.state,parents

    currmat = currnode.state

    # now we need to make the next nodes
    pos1 = []

    for i in range(3):
      for j in range(3):
        if(currmat[i][j] == key):
            pos1.append(i)
            pos1.append(j)
            break


    # print("Now doing state")
    # print(currmat)
    # print("\n\n")

    # print(*pos1)

    if(pos1[0]>0):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 

      nextmat = swap(newmat, [pos1[0]-1, pos1[1]], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        frontier.append(newnode1)

    if(pos1[0]<2):

      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0]+1, pos1[1]], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        frontier.append(newnode1) 

    if(pos1[1]>0):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0], pos1[1]-1], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'

        parents[temp] = currt

        frontier.append(newnode1) 

    if(pos1[1]<2):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0], pos1[1]+1], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):  
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        frontier.append(newnode1)

    # print("\n")
    # print(*currmat)
    temp = tuple(map(tuple, currmat))
    explored.add(temp)
    
```
