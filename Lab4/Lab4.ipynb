```
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```
def finderror(centres, points):

  ferror = 0.0
  for i in range(len(centres)):
     if(len(points[i])>0):
      ferror = ferror + (np.linalg.norm(centres[i] - points[i]))**2
  
  return ferror
```

```
def plotpoints(newsets):
  colors = ["red","green","blue","yellow","black","orange","purple","brown","gray","cyan","magenta"]
  count = 0
  for i in newsets:
    if(len(i)==0):
      continue
    j = np.array(i)
    plt.scatter(j[:,0], j[:,1], c = colors[count])
    count+=1
```

```
def plotcentres(centres, k):
  colors = ["red","green","blue","yellow","black","orange","purple","brown","gray","cyan","magenta"]
  plt.scatter(centres[:,0], centres[:,1], c = colors[0:k], marker = '*',s= 1000)
```

```
def finalkmeans(X, centres, k):

  pos = True 
  n = len(X)
  sse = []

  while(pos):

    newcentres = []
    newsets = []

    for i in range(k):
      arr = []
      newsets.append(arr)

    distances = []

    for i in range(n):
      arr1 = np.linalg.norm(X[i] - centres, axis = 1)
      distances.append(arr1)
    
    for i in range(n):

      index = np.argmin(distances[i])
      newsets[index].append(X[i])
    
    newsets = np.array(newsets)


    for i in range(k):

      if(len(newsets[i])>0):
        temp = np.array(newsets[i])
        tempmean = np.mean(temp, axis =0)
        newcentres.append(tempmean.tolist())

      else:
        newcentres.append(centres[i])

    newcentres = np.array(newcentres)
    centres = np.array(centres)

    comparison = newcentres == centres 

    # colors = ["red", "blue", "green"]
    # plt.scatter(centres[:,0], centres[:,1], c = colors[0:k], marker = '*',s= 1000)

    ssetemp = finderror(centres, newsets)
    sse.append(ssetemp)
    if(comparison.all()):
      fcentres = centres
      pos = False 
    
    centres = newcentres

  plotcentres(centres, k)
  plotpoints(newsets)
  plt.show()

  return fcentres, newsets, sse
```
