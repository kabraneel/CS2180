```
n = int(input())
tx = np.random.rand(n,2)
```

```
plt.scatter(tx[:,0],tx[:,1])
```

```
def compute(t,new):
  return np.linalg.norm(t-new)
```

```
t2x = np.random.uniform(-1,1,(n,2))
```

```
x,y = input("Enter x and y coordinates of the point ").split()

new = np.array([float(x),float(y)])
d = []
for i in t2x:
    d.append([compute(new,i), i])

d.sort()
points = []
# k will take values 5,10,15
for k in [5,10,15]:
    for i in range(k):
        print(d[i][1])
        if(k == 15):
          points.append(d[i][1])
    print("\n\n")
points = np.array(points)
```

```
def findangles(v):
    b = []
    for i in t2x:
        if(i[0]*v[0] + i[1]*v[1]>0):
          b.append(i)
    return b
```

```
angles = findangles(new)
toplot = []
for k in [5,10,15]:
    for i in range(k):
        if i<len(angles):
          if(k == 15):
            toplot.append(angles[i])
          print(angles[i])
    print("\n\n")

toplot = np.array(toplot)
```

```
array = np.random.uniform(-0.5, 0.5, 100)
k = [i+1 for i in range(100)]
k = np.array(k)

y = []

for i in k:
  y.append(np.mean(array[:i]))

plt.plot(k,y, linewidth = '5')
# plot of sqrt(1/k)
plt.plot(k, (1/k)**0.5, linewidth = '5')
#plot of -1*sqrt(1/k)
plt.plot(k, -1*((1/k)**0.5), linewidth = '5')
plt.show()
```
