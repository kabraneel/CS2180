```python
import numpy as np
```

```python
pmin = np.array([0.01, 0.02, 0.03, 0.1, 0.3])
pmax = np.array([0.1, 0.2, 0.3, 0.5, 0.7])
shots = np.array([1,2,3,4,6]) 
```

```python
def pout(w,a):
  return (pmax[a] +(pmin[a]-pmax[a])*(w-1)/9) 
```

```python
# fill the initial vectors with 0
dp = np.zeros((301,11))
actions = np.zeros((301,11))


# zero wickets
for i in range(301):
  dp[i][0] = 0
  actions[i][0] = -1

# 1 ball left
for i in range(1,11):
  
  # mexval denotes the maximum value over all actions
  mexval = -1
  # mexind denotes the actoin which leads to the max value
  mexind = -1 
  for j in range(5):

    if((1 - pout(i,j))*shots[j]> mexval):
      mexval = (1 - pout(i,j))*shots[j] 
      mexind = shots[j] 
    
  dp[1][i] = mexval 
  actions[0][i] = mexind



for i in range(2,301):
  for j in range(1,11):
    mexval = -1
    mexind = -1

    for k in range(5):

      if(((1 - pout(j,k))*(shots[k] + dp[i-1][j]) + pout(j,k)*(dp[i-1][j-1]))>mexval):

        mexval = ((1 - pout(j,k))*(shots[k] + dp[i-1][j]) + pout(j,k)*(dp[i-1][j-1])) 
        mexind = shots[k] 

    dp[i][j] = mexval 
    actions[i][j] = mexind 

# actions contain the action at w,b
print(dp[300][10])
print(*actions)
```

```python
for i in dp:
  for x in i:
    print(round(x,2), end = "  ")
  print("\n")
```
