```python
def findcost1(mat):

  goals = [[[1,2,3], [4,5,6], [7,'a','b']], [[1,2,3], [4,5,6], [7,'b','a']]] 
  cost = 1000000
  n = len(mat)
  for k in goals:
    tcost = 0
    # compute the cost 
    for i in range(n):
      for j in range(n):
        x = mat[i][j] 

        for i1 in range(n):
          for j1 in range(n):
            if(k[i1][j1] == x):
              tcost += abs(i1-i) + abs(j1-j) 
              break 
    if(tcost<cost):
      cost = tcost 

  return cost
```

```python
class Node1:

  def __init__(self):
    self.state = []
    self.parent = []
    self.todo = 'a'
    self.cost = 0
    self.final = -1
    self.h = -1
    
  def __lt__(self, other):

    if(self.final<other.final):
      return True
    return False
```

```python
def swap(mat,ele1, ele2):

  mat1 = []
  for i in mat:
    mat1.append(i)

  temp = mat1[ele1[0]][ele1[1]] 
  mat1[ele1[0]][ele1[1]] = mat1[ele2[0]][ele2[1]]
  mat1[ele2[0]][ele2[1]] = temp

  return mat1
```

```python
def solve1(start):

  parents = {}
  
  states = []
  frontier = []

  goalstates = [[[1,2,3], [4,5,6], [7,'a','b']], [[1,2,3], [4,5,6], [7,'b','a']]] 
  heapq.heapify(frontier)
  explored = set()

  newnode = Node1()
  newnode.state = start
  newnode.todo = 'a'
  newnode.h = findcost1(start)
  newnode.cost = 0
  newnode.final = newnode.h + newnode.cost

  heapq.heappush(frontier,newnode)

  count = 0

  parents[tuple(map(tuple, start))] = None

  while(len(frontier)>0):

    count+=1
    currnode = heapq.heappop(frontier)
    print("cost == ", currnode.final)
    # if(count == 10):
    #   return currnode.state,parents
    currt = tuple(map(tuple, currnode.state))
    
    key = currnode.todo
    if currt in explored:
      continue 

    states.append(currnode)

    if currnode.state in  goalstates:
      return currnode.state,parents

    currmat = currnode.state

    # now we need to make the next nodes
    pos1 = []

    for i in range(3):
      for j in range(3):
        if(currmat[i][j] == key):
            pos1.append(i)
            pos1.append(j)
            break


    # print("Now doing state")
    # print(currmat)
    # print("\n\n")

    # print(*pos1)

    if(pos1[0]>0):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 

      nextmat = swap(newmat, [pos1[0]-1, pos1[1]], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        newnode1.cost = currnode.cost + 2 
        newnode1.h = findcost1(newnode1.state)
        newnode1.final = newnode1.cost + newnode1.h
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        heapq.heappush(frontier,newnode1)

    if(pos1[0]<2):

      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0]+1, pos1[1]], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        newnode1.cost = currnode.cost + 2 
        newnode1.h = findcost1(newnode1.state)
        newnode1.final = newnode1.cost + newnode1.h
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        heapq.heappush(frontier,newnode1)

    if(pos1[1]>0):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0], pos1[1]-1], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        newnode1.cost = currnode.cost + 2 
        newnode1.h = findcost1(newnode1.state)
        newnode1.final = newnode1.cost + newnode1.h
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'

        parents[temp] = currt

        heapq.heappush(frontier,newnode1)

    if(pos1[1]<2):
      
      newmat = []
      for i in currmat:
        newmat.append(i.copy()) 
      
      nextmat = swap(newmat, [pos1[0], pos1[1]+1], pos1)
      temp = tuple(map(tuple, nextmat))
      if(temp not in explored):  
        newnode1 = Node1()
        newnode1.state = nextmat
        newnode1.parent = currmat
        newnode1.cost = currnode.cost + 2
        newnode1.h = findcost1(newnode1.state)
        newnode1.final = newnode1.cost + newnode1.h
        if(key == 'a'):
          newnode1.todo = 'b'
        else:
          newnode1.todo = 'a'
        parents[temp] = currt
        heapq.heappush(frontier,newnode1)

    # print("\n")
    # print(*currmat)
    temp = tuple(map(tuple, currmat))
    explored.add(temp)
```
