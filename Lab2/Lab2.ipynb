**QUESTION 2**

Code to create the intial environment

---
Exactly 10 positions contain a value of dirt between (0,100)


---

We will print the grid after producing it





```
xsize = 20
ysize = 20

grid = np.full((xsize,ysize), 0)
count = 0
while(count<10):
  x = np.random.randint(0,xsize,size = (2))
  # print(x[0])
  if(grid[x[0]][x[1]] != 0):
    continue
  dirt = np.random.randint(100)
  grid[x[0]][x[1]] = dirt
  count += 1

print(grid)
```

Now code to see what happens to the enviornment

---
Input: Operation, Current state, Output: Next state, reward


```
#this function takes in the current state
#and based on the operation returns the next state
def environment(op, xt, yt):

  #values of op :
  # 0 -> Left
  # 1 -> Right
  # 2 -> Up
  # 3 -> Down
  # 4 -> pick-dirt
  if(op==0):
    if(xt>0):
      return [xt-1,yt, 0]
    
    else:
      return [xt,yt, -10]


  if(op == 1):
    
    if(xt<xsize-1):
      return [xt+1, yt, 0]
    
    else:
      return [xt, yt, -10]
    

  if(op==2):
    
    if(yt>0):
      return [xt,yt-1,0]
    
    else:
      return [xt,yt,-10]


  if(op == 3):
    
    if(yt<ysize-1):
      return [xt, yt+1, 0]
    
    else:
      return [xt, yt, -10]


  else:

    if(grid[xt][yt] == 0):
      return [xt,yt, -1]
    
    else:
      temp = grid[xt][yt]
      grid[xt][yt] = 0
      return [xt, yt, temp]
```

Now to create the agent code

---
Input: Current state, Output: Operation


```
def agent(position):

  op = np.random.randint(5)
  return op
```

Now the driver code


```
operations = ["Left", "Right", "Up", "Down", "Pick-dirt"]
t = 100

position = np.random.randint(xsize,size = 2)
print("Initial position", *position)

reward = 0

print(grid)

for i in range(t):

  operation = agent(position)

  arr = environment(operation,position[0],position[1])

  position = arr[0:2]
  statereward = arr[2]

  print(i+1, *position, operations[operation], statereward)

  reward = reward + statereward




```

```
#Code for average reward and printing the final grid
print(reward/t)
print(grid)
```
